```
# Import libraries 
import matplotlib.pyplot as plt 
import numpy as np 
# define the function 
def f(x):
  return 2 * x * x * np.cos(x) - 5 * x
# Creating vectors X and Y 
x = np.linspace(-5, 5, 1000) #numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0) The starting value of the sequence,The end value of the sequence,Number of samples to generate. Default is 50. Must be non-negative.
y = f(x)
  
fig = plt.figure(figsize = (10, 5)) 
# Create the plot 
plt.plot(x, y) 
#print a value of x
print(f(-1))
# Show the plot 
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
#initial point x0
x = [-1.]
fx=[]
def df(x):
    return 4 * x * np.cos(x) - 2 * x * x * np.sin(x) - 5
alpha = 0.1#alpha = 0.05
for i in range(0,1000):
  x.append(x[i] - alpha * df(x[i]))
  fx.append(f(x[i]))
  plt.scatter(x[i], f(x[i]))

print(min(fx))
plt.show()  
```

```
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import random

def generateSample(N, variance=100):
    X = np.matrix(range(N)).T + 1
    Y = np.matrix([random.random() * variance + i * 10 + 900 for i in range(len(X))]).T
    return X, Y

def fitModel_gradient(x, y):
    N = len(x)
    w = np.zeros((x.shape[1], 1))
    eta = 0.0001

    maxIteration = 100000
    for i in range(maxIteration):
        error = x * w - y
        gradient = x.T * error / N
        w = w - eta * gradient
    return w

def plotModel(x, y, w):
    plt.plot(x[:,1], y, "x")
    plt.plot(x[:,1], x * w, "r-")
    plt.show()

def test(N, variance, modelFunction):
    X, Y = generateSample(N, variance)
    X = np.hstack([np.matrix(np.ones(len(X))).T, X])
    w = modelFunction(X, Y)
    plotModel(X, Y, w)

test(50, 600, fitModel_gradient)
test(50, 1000, fitModel_gradient)
test(100, 200, fitModel_gradient)
```
